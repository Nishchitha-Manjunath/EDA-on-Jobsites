```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
joblist = []

for x in range(0,100):
  url = f'https://www.indeed.com/jobs?q=data+scientist&start='
  r = requests.get(url + str(x))
  soup = BeautifulSoup(r.content,'html.parser')
  content = soup.find_all('div', class_ = 'jobsearch-SerpJobCard')

  for item in content:
    title = item.find('a').text.strip()  #try strip
    #company = item.find('span', class_ = 'company').text.strip() ignored it to match other csv
    #print(company) #testing of company test-4
    try:
      salary = item.find('span', class_ = 'salaryText').text.strip()
    except:
      salary = ''
    jobtype = getattr(item.find('span', class_ = 'remote'), 'text', None)
    description = item.find('div', class_ = 'summary').text.strip().replace('\n', '')
    location = getattr(item.find('div', class_ = 'location accessible-contrast-color-location'), 'text', None)
    
    #location = item.find('span', class_ = 'location accessible-contrast-color-location').text.strip()
    #print(location) 
    job = {
        'title' : title,
        #'company' : company,
        'salary' : salary,
        'jobtype' : jobtype,
        'description' : description,
        'location' : location
    }
    joblist.append(job)  


  print('jobs Found :', len(joblist))  
  time.sleep(2)
print(joblist)


df = pd.DataFrame(joblist)

print(df.head())

df.to_csv('indeed_jobs.csv')
```
